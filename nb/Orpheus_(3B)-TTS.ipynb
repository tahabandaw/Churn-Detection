```python
import pandas as pd
import io

# 1. --- SETUP: Create a sample DataFrame mimicking your data ---
csv_data = "/content/Upwork_Taha/metadata.csv"

df = pd.read_csv(csv_data)


# 2. --- DEFINE: Create the complete standardization function ---
def standardize_transcription(text):
    """
    Cleans, standardizes, and ensures proper end-of-sentence punctuation
    for Haitian-Creole transcriptions.
    """
    # Ensure the input is a string and remove leading/trailing whitespace
    text = str(text).strip()

    # Split the text into words to safely check the first word
    words = text.split()

    # Return the original text if it's empty
    if not words:
        return text

    # --- Rule 1: Handle Questions ---
    # Check for question words at the beginning of the sentence
    if words[0].lower() in ['kijan', 'eske', '√®ske']:
        # Correct the spelling
        if words[0].lower() == 'kijan':
            words[0] = 'K√≤man'
        else:
            words[0] = '√àske'

        updated_text = ' '.join(words)
        # Remove any existing punctuation and add a question mark
        return updated_text.rstrip('.,?! ') + '?'

    # --- Rule 2: Handle Statements and Commands (Default) ---
    else:
        # For all other sentences, we assume they are statements or neutral commands.
        # First, remove any potentially incorrect or missing end punctuation.
        updated_text = text.rstrip('.,?! ')

        # Now, add a period.
        return updated_text + '.'


# 3. --- APPLY: Run the function on the 'transcription' column ---
print("----------- BEFORE STANDARDIZATION -----------")
print(df)
print("\n" + "="*50 + "\n")

df['transcription'] = df['transcription'].apply(standardize_transcription)

print("----------- AFTER STANDARDIZATION -----------")
print(df[['transcription', 'transcription']])

# To save, you would select the final columns you need
# final_df = df[['file_name', 'standardized_transcription', 'speaker_id']]
# final_df.to_csv('cleaned_dataset.csv', index=False, header=['file_name', 'transcription', 'speaker_id'])
```

```python
import pandas as pd
import subprocess
from pathlib import Path
from tqdm.auto import tqdm # A library to show a beautiful progress bar

# ===================================================================
#
#                         CONFIGURATION
#
# ===================================================================

# 1. --- Define the paths ---
METADATA_CSV_PATH = 'cleaned_dataset.csv'
ORIGINAL_AUDIO_BASE_DIR = Path('/content/Upwork_Taha')
CLEANED_AUDIO_DIR = Path('/content/cleaned_audio')


# ===================================================================
#
#                AUDIO CLEANING FUNCTION (using FFmpeg)
#
# ===================================================================

def clean_one(src_path: Path, dest_path: Path):
    """
    Applies a series of FFmpeg filters to clean and normalize a single audio file.
    """
    # Create the parent directory for the destination file if it doesn't exist
    dest_path.parent.mkdir(parents=True, exist_ok=True)

    filt = (
        "highpass=f=80,"
        "afftdn,"
        "loudnorm=I=-16:LRA=11:TP=-1.5,"
        "dynaudnorm=f=200,"
        "apad=pad_dur=0.1"
    )

    try:
        command = [
            "ffmpeg", "-i", str(src_path), "-af", filt,
            "-ar", "24000", "-ac", "1", "-y", "-hide_banner", "-loglevel", "error",
            str(dest_path)
        ]
        subprocess.run(command, check=True, capture_output=True, text=True)
        return True, None

    except FileNotFoundError:
        error_message = f"ERROR: Source file not found at {src_path}"
        return False, error_message

    except subprocess.CalledProcessError as e:
        error_message = f"ERROR: FFmpeg failed on {src_path}.\nFFmpeg error:\n{e.stderr}"
        return False, error_message


# ===================================================================
#
#                          MAIN SCRIPT
#
# ===================================================================

# --- 1. Load the metadata ---
print(f"Loading metadata from '{METADATA_CSV_PATH}'...")
try:
    df = pd.read_csv(METADATA_CSV_PATH)
    print(f"Successfully loaded {len(df)} records.")
except FileNotFoundError:
    print(f"FATAL ERROR: The file '{METADATA_CSV_PATH}' was not found.")
    exit()

# =================================================================================
#
#  =====> THE FIX IS HERE <=====
#  We remove the unwanted 'Audio/' prefix from the 'file_name' column.
#
# =================================================================================
print("\nVerifying and cleaning file paths from the CSV...")
# First, let's see what the path looks like before the fix
print("Path example BEFORE cleaning:", df['file_name'].iloc[0])

# Now, remove the prefix
df['file_name'] = df['file_name'].str.removeprefix('Audio/')

# And let's see what it looks like after
print("Path example AFTER cleaning:", df['file_name'].iloc[0])


# --- 2. Create the destination directory ---
print(f"\nCreating destination directory for cleaned audio: '{CLEANED_AUDIO_DIR}'")
CLEANED_AUDIO_DIR.mkdir(parents=True, exist_ok=True)


# --- 3. Process each audio file ---
print("\nStarting audio cleaning process...")
new_file_paths = []
failed_files = []

for index, row in tqdm(df.iterrows(), total=df.shape[0], desc="Cleaning Audio"):
    original_relative_path = row['file_name']

    # This path will now be correct: /content/Upwork_Taha/509449...
    source_path = ORIGINAL_AUDIO_BASE_DIR / original_relative_path

    # We now need to create subdirectories in the cleaned_audio folder
    # to avoid having all 18,000 files in one place.
    cleaned_relative_path = Path(original_relative_path)
    cleaned_full_path = CLEANED_AUDIO_DIR / cleaned_relative_path

    success, error = clean_one(source_path, cleaned_full_path)

    if success:
        new_file_paths.append(str(cleaned_relative_path)) # Save the relative path
    else:
        print(error)
        failed_files.append(source_path)
        new_file_paths.append(None)


# --- 4. Update the DataFrame and save the new CSV ---
print("\nCleaning process finished.")
print(f"Successfully processed: {len(df) - len(failed_files)} files.")
print(f"Failed to process: {len(failed_files)} files.")

df['file_name'] = new_file_paths
df.dropna(subset=['file_name'], inplace=True)

output_csv_path = 'metadata_cleaned.csv'
df.to_csv(output_csv_path, index=False)

print(f"\nUpdated metadata saved to '{output_csv_path}'.")
print("This new CSV file points to the high-quality audio in the 'cleaned_audio' directory.")
```

```python
from datasets import load_dataset, Features, Value, Audio
import os # Import the 'os' module to handle file paths

# ========================================================================
# --- CONFIGURE THIS VARIABLE ---
# Set this to the absolute path of the directory that CONTAINS your 'Audio' folder.
# For example, if your files are in '/home/user/my_project/Audio/...',
# then set base_path to '/home/user/my_project/'
#
# If your notebook is in the same folder as the "Audio" folder, you can often just use "."
base_path = "/content/cleaned_audio" # <--- IMPORTANT: CHANGE THIS
# ========================================================================


# Define the path to your metadata CSV file
metadata_file = os.path.join(base_path, "/content/cleaned_audio/metadata_cleaned.csv") # Assumes CSV is in the base path too

# We define the exact data type for each column in your CSV.
feature_types = Features({
    "file_name": Value("string"),
    "transcription": Value("string"),
    "speaker_id": Value("string"),
    "gender": Value("string"),
    "accent": Value("string"),
    "score": Value("float"),
    "duration": Value("float")
})

# Load the dataset using our defined feature types
dataset = load_dataset(
    'csv',
    data_files=metadata_file,
    features=feature_types,
    split='train'
)

# --- THE FIX IS HERE ---
# The paths in your CSV are relative. We need to create the full path.
# This function takes a row, gets the relative path from 'file_name',
# and joins it with your 'base_path' to create the full, correct path.
def create_absolute_path(example):
    example["file_name"] = os.path.join(base_path, example["file_name"])
    return example

# Apply this function to every row in the dataset
dataset = dataset.map(create_absolute_path)


# --- The rest of the script continues as before ---

# 1. Rename columns to match the model's expectations
dataset = dataset.rename_column("transcription", "text")
dataset = dataset.rename_column("speaker_id", "source")

# 2. Load the audio files from the now-correct file paths
dataset = dataset.cast_column("file_name", Audio())
dataset = dataset.rename_column("file_name", "audio")

# Now, your 'dataset' object is correctly formatted and ready for tokenization.
print("Dataset loaded and paths corrected successfully!")
print("First example:", dataset[0])

```

```python
#@title Tokenization Function

import locale
import torchaudio.transforms as T
import os
import torch
from snac import SNAC
locale.getpreferredencoding = lambda: "UTF-8"
ds_sample_rate = dataset[0]["audio"]["sampling_rate"]

snac_model = SNAC.from_pretrained("hubertsiuzdak/snac_24khz")
snac_model = snac_model.to("cuda")
def tokenise_audio(waveform):
  waveform = torch.from_numpy(waveform).unsqueeze(0)
  waveform = waveform.to(dtype=torch.float32)
  resample_transform = T.Resample(orig_freq=ds_sample_rate, new_freq=24000)
  waveform = resample_transform(waveform)

  waveform = waveform.unsqueeze(0).to("cuda")

  #generate the codes from snac
  with torch.inference_mode():
    codes = snac_model.encode(waveform)

  all_codes = []
  for i in range(codes[0].shape[1]):
    all_codes.append(codes[0][0][i].item()+128266)
    all_codes.append(codes[1][0][2*i].item()+128266+4096)
    all_codes.append(codes[2][0][4*i].item()+128266+(2*4096))
    all_codes.append(codes[2][0][(4*i)+1].item()+128266+(3*4096))
    all_codes.append(codes[1][0][(2*i)+1].item()+128266+(4*4096))
    all_codes.append(codes[2][0][(4*i)+2].item()+128266+(5*4096))
    all_codes.append(codes[2][0][(4*i)+3].item()+128266+(6*4096))


  return all_codes

def add_codes(example):
    # Always initialize codes_list to None
    codes_list = None

    try:
        answer_audio = example.get("audio")
        # If there's a valid audio array, tokenise it
        if answer_audio and "array" in answer_audio:
            audio_array = answer_audio["array"]
            codes_list = tokenise_audio(audio_array)
    except Exception as e:
        print(f"Skipping row due to error: {e}")
        # Keep codes_list as None if we fail
    example["codes_list"] = codes_list

    return example

dataset = dataset.map(add_codes, remove_columns=["audio"])

tokeniser_length = 128256
start_of_text = 128000
end_of_text = 128009

start_of_speech = tokeniser_length + 1
end_of_speech = tokeniser_length + 2

start_of_human = tokeniser_length + 3
end_of_human = tokeniser_length + 4

start_of_ai = tokeniser_length + 5
end_of_ai =  tokeniser_length + 6
pad_token = tokeniser_length + 7

audio_tokens_start = tokeniser_length + 10

dataset = dataset.filter(lambda x: x["codes_list"] is not None)
dataset = dataset.filter(lambda x: len(x["codes_list"]) > 0)

def remove_duplicate_frames(example):
    vals = example["codes_list"]
    if len(vals) % 7 != 0:
        raise ValueError("Input list length must be divisible by 7")

    result = vals[:7]

    removed_frames = 0

    for i in range(7, len(vals), 7):
        current_first = vals[i]
        previous_first = result[-7]

        if current_first != previous_first:
            result.extend(vals[i:i+7])
        else:
            removed_frames += 1

    example["codes_list"] = result

    return example

dataset = dataset.map(remove_duplicate_frames)

tok_info = '''*** HERE you can modify the text prompt
If you are training a multi-speaker model (e.g., canopylabs/orpheus-3b-0.1-ft),
ensure that the dataset includes a "source" field and format the input accordingly:
- Single-speaker: f"{example['text']}"
- Multi-speaker: f"{example['source']}: {example['text']}"
'''
print(tok_info)

def create_input_ids(example):
    # Determine whether to include the source field
    text_prompt = f"{example['source']}: {example['text']}" if "source" in example else example["text"]

    text_ids = tokenizer.encode(text_prompt, add_special_tokens=True)
    text_ids.append(end_of_text)

    example["text_tokens"] = text_ids
    input_ids = (
        [start_of_human]
        + example["text_tokens"]
        + [end_of_human]
        + [start_of_ai]
        + [start_of_speech]
        + example["codes_list"]
        + [end_of_speech]
        + [end_of_ai]
    )
    example["input_ids"] = input_ids
    example["labels"] = input_ids
    example["attention_mask"] = [1] * len(input_ids)

    return example


dataset = dataset.map(create_input_ids, remove_columns=["text", "codes_list"])
columns_to_keep = ["input_ids", "labels", "attention_mask"]
columns_to_remove = [col for col in dataset.column_names if col not in columns_to_keep]

dataset = dataset.remove_columns(columns_to_remove)
```

```python
prompts = [
    "Nan yon ti vilaj an Ayiti ki chaje ak koul√®, kote pye kokoye danse nan briz lanm√® a, te gen yon kominote ki te gen anpil k√® kontan. Chak maten, son tanbou te melanje ak ri timoun ki t ap kouri sou wout an t√®. Machann yo te etale b√®l fwi twopikal yo, tankou mango ak anana ki te santi bon. Moun yo te toujou pataje sa yo te genyen, menm si se te ti kras. F√≤s p√®p nou an se nan solidarite sa a li chita, yon eritaj zans√®t nou yo kite pou nou.",
]

chosen_voice = "agnusdeilouisgabriel@gmail.com" # None for single-speaker
```

```python
#@title Run Inference


FastLanguageModel.for_inference(model) # Enable native 2x faster inference

# Moving snac_model cuda to cpu
snac_model.to("cpu")

prompts_ = [(f"{chosen_voice}: " + p) if chosen_voice else p for p in prompts]

all_input_ids = []

for prompt in prompts_:
  input_ids = tokenizer(prompt, return_tensors="pt").input_ids
  all_input_ids.append(input_ids)

start_token = torch.tensor([[ 128259]], dtype=torch.int64) # Start of human
end_tokens = torch.tensor([[128009, 128260]], dtype=torch.int64) # End of text, End of human

all_modified_input_ids = []
for input_ids in all_input_ids:
  modified_input_ids = torch.cat([start_token, input_ids, end_tokens], dim=1) # SOH SOT Text EOT EOH
  all_modified_input_ids.append(modified_input_ids)

all_padded_tensors = []
all_attention_masks = []
max_length = max([modified_input_ids.shape[1] for modified_input_ids in all_modified_input_ids])
for modified_input_ids in all_modified_input_ids:
  padding = max_length - modified_input_ids.shape[1]
  padded_tensor = torch.cat([torch.full((1, padding), 128263, dtype=torch.int64), modified_input_ids], dim=1)
  attention_mask = torch.cat([torch.zeros((1, padding), dtype=torch.int64), torch.ones((1, modified_input_ids.shape[1]), dtype=torch.int64)], dim=1)
  all_padded_tensors.append(padded_tensor)
  all_attention_masks.append(attention_mask)

all_padded_tensors = torch.cat(all_padded_tensors, dim=0)
all_attention_masks = torch.cat(all_attention_masks, dim=0)

input_ids = all_padded_tensors.to("cuda")
attention_mask = all_attention_masks.to("cuda")
generated_ids = model.generate(
      input_ids=input_ids,
      attention_mask=attention_mask,
      max_new_tokens=1200,
      do_sample=True,
      temperature=0.6,
      top_p=0.95,
      repetition_penalty=1.1,
      num_return_sequences=1,
      eos_token_id=128258,
     use_cache = True
  )
token_to_find = 128257
token_to_remove = 128258

token_indices = (generated_ids == token_to_find).nonzero(as_tuple=True)

if len(token_indices[1]) > 0:
    last_occurrence_idx = token_indices[1][-1].item()
    cropped_tensor = generated_ids[:, last_occurrence_idx+1:]
else:
    cropped_tensor = generated_ids

mask = cropped_tensor != token_to_remove

processed_rows = []

for row in cropped_tensor:
    masked_row = row[row != token_to_remove]
    processed_rows.append(masked_row)

code_lists = []

for row in processed_rows:
    row_length = row.size(0)
    new_length = (row_length // 7) * 7
    trimmed_row = row[:new_length]
    trimmed_row = [t - 128266 for t in trimmed_row]
    code_lists.append(trimmed_row)


def redistribute_codes(code_list):
  layer_1 = []
  layer_2 = []
  layer_3 = []
  for i in range((len(code_list)+1)//7):
    layer_1.append(code_list[7*i])
    layer_2.append(code_list[7*i+1]-4096)
    layer_3.append(code_list[7*i+2]-(2*4096))
    layer_3.append(code_list[7*i+3]-(3*4096))
    layer_2.append(code_list[7*i+4]-(4*4096))
    layer_3.append(code_list[7*i+5]-(5*4096))
    layer_3.append(code_list[7*i+6]-(6*4096))
  codes = [torch.tensor(layer_1).unsqueeze(0),
         torch.tensor(layer_2).unsqueeze(0),
         torch.tensor(layer_3).unsqueeze(0)]

  # codes = [c.to("cuda") for c in codes]
  audio_hat = snac_model.decode(codes)
  return audio_hat

my_samples = []
for code_list in code_lists:
  samples = redistribute_codes(code_list)
  my_samples.append(samples)
from IPython.display import display, Audio
if len(prompts) != len(my_samples):
  raise Exception("Number of prompts and samples do not match")
else:
  for i in range(len(my_samples)):
    print(prompts[i])
    samples = my_samples[i]
    display(Audio(samples.detach().squeeze().to("cpu").numpy(), rate=24000))
# Clean up to save RAM
del my_samples,samples
```
